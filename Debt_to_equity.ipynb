```python
# Drop column 'Year'
data = data.drop('Year', axis=1)
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler() 
data[['Total_Liabilities', 'Long_Term_Debt', 'Short_Term_Debt', 'Total_Assets', 'Total_Equity', 'Net_Income', 'Operating_Cash_Flow', 'Current_Ratio', 'Fixed_Assets', 'Retained_Earnings']] = scaler.fit_transform(data[['Total_Liabilities', 'Long_Term_Debt', 'Short_Term_Debt', 'Total_Assets', 'Total_Equity', 'Fixed_Assets', 'Operating_Cash_Flow', 'Current_Ratio', 'Net_Income', 'Retained_Earnings']]) 
data.head()
```

SELECT FEATURES AND TARGET


```python
x = data.drop('Debt_to_Equity', axis=1)
y = data['Debt_to_Equity']   
```

GETTING TRAING SET AND TEST SET

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)    
print("x_train shape:", x_train.shape)
print("x_test shape:", x_test.shape)
```

BUILD MODEL

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import xgboost as xgb
```

```python
model = xgb.XGBRegressor(
 n_estimators=300,
 learning_rate=0.05,
 max_depth=5, # maximum depth of each tree
  subsample=0.8,
  colsample_bytree=0.8,
   random_state=42 
)
model.fit(x_train, y_train)
```

```python
# Make predictions
y_pred = model.predict(x_test)
```

```python
# Evaluate model performance
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("----- XGBOOST REGRESSION PERFORMANCE -----")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R-squared (R²): {r2:}")
```

VISUALIZE ACTUAL VS PREDICTED

```python
plt.figure(figsize=(6, 5))
sns.scatterplot(x=y_test, y=y_pred, color='royalblue')
plt.xlabel("Actual Debt_to_Equity")
plt.ylabel("Predicted Debt_to_Equity")
plt.title("Actual vs Predicted - XGBoost Regression")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')  # perfect prediction line
plt.show()
```

CROSS VALIDATION

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, x, y, scoring='r2', cv=5)
print("Cross-validated R² scores:", scores)
print("Mean R²:", scores.mean())
```
