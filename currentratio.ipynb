```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler() 
data[['cash_and_equivalents', 'accounts_receivable', 'inventory', 'prepaid_expenses', 'marketable_securities', 'accounts_payable', 'short_term_loans', 'accrued_expenses', 'current_portion_long_term_debt', 'operating_cash_flow', 'net_sales']] = scaler.fit_transform(data[['cash_and_equivalents', 'accounts_receivable', 'inventory', 'prepaid_expenses', 'marketable_securities', 'accounts_payable', 'short_term_loans', 'accrued_expenses', 'current_portion_long_term_debt', 'operating_cash_flow', 'net_sales']]) 
data.head()
```

SELECT FEATURES AND TARGET



```python
x = data.iloc[:, :-1]
y = data.iloc[:, -1]    
```

GETTING A TRAINING SET AND TEST SET


```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)    
print("x_train shape:", x_train.shape)
print("x_test shape:", x_test.shape)
```

RESHAPE X_TRAIN TO 2D

```python
if len(x_train.shape) == 1:
    X_train = np.expand_dims(x_train, axis=1)  
```

# Convert X_train to a NumPy array, ,, #Convert X_test to a NumPy array, # Ensure y_train is a 1D array and # Ensure y_test to a Numpy array

```python
# Convert x_train to a NumPy array
x_train = x_train.values if isinstance(x_train, pd.DataFrame) else x_train
#Convert x_test to a NumPy array
x_test = x_test.values if isinstance(x_test, pd.DataFrame) else x_test
# Ensure y_train is a 1D array
y_train = y_train.values.ravel() if isinstance(y_train, pd.Series) else y_train
# Ensure y_test to a Numpy array
y_test = y_test.values if isinstance(y_test, pd.DataFrame) else y_test
```

BUILDING THE MODEL

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

MAKING PREDICTION

```python
y_pred = model.predict(x_test)
print("Predicted values:", y_pred)
```

EVALUATE MODEL ACCURACY


```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("----- MODEL PERFORMANCE -----")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.2f}")
print(f"R-squared (R²): {r2:.2f}")
```

Visualize Actual vs Predicted

```python
plt.figure(figsize=(6, 5))
sns.scatterplot(x=y_test, y=y_pred, color='royalblue')
plt.xlabel("Actual current ratio")
plt.ylabel("Predicted current ratio")
plt.title("Actual vs Predicted - Linear Regression")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')  # perfect prediction line
plt.show()
```

Check Coefficients

```python
coeff_df = pd.DataFrame({
    'Feature': x.columns,
    'Coefficient': model.coef_
})
print("\n----- FEATURE COEFFICIENTS -----")
print(coeff_df)
```

Cross-Validation

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, x, y, scoring='r2', cv=5)
print("Cross-validated R² scores:", scores)
print("Mean R²:", scores.mean())
```
